```python
from datasets import load_dataset
import json
import os
import glob
```

```python
dataset = load_dataset("reglab/barexam_qa", name="qa", trust_remote_code=True)
```

```python
val_data = dataset["validation"]
```

```python
print(json.dumps(val_data[0], indent=4))
```

```python
idx = set()
for question in val_data:
    idx.add(question["idx"])

len(idx)
```

```python
answer_dict = {}
for question in val_data:
    answer_dict[question["idx"]] = question["answer"]
```

```python
print(answer_dict)
```

Get the answers from the experiment json files

```python
lmars_answers_path = "..//..//Experiments//bar_exam/L-MARS/SimpleWorkflow/results"
```

```python
lmars_answer_dict = {}
for filename in glob.glob(os.path.join(lmars_answers_path, '*.json')):
    with open(filename, mode='r', encoding='utf-8') as currentFile:
        data = json.load(currentFile)
        passage_idx = data.get("passage_idx")
        answer = data["final_result"]["final_answer"]["data"]["answer"]
        lmars_answer_dict[passage_idx] = answer
```

```python
lmars_answer_dict
```

```python
for key, val in lmars_answer_dict.items():
    if len(val) > 1:
        print(key)
```

```python
correct = 0
for key, answer in lmars_answer_dict.items():
    true_answer = answer_dict.get(key)
    if answer == true_answer:
        correct += 1

print("The accuracy is: ", correct/len(lmars_answer_dict))
```

Accuracy for LMARS on barexam_qa is 0.65

```python
path = "..//..//Experiments//bm25"
bm25_answer_dict = {}
for filename in glob.glob(os.path.join(path, '*.json')):
    print(filename)
    with open(filename, mode='r', encoding='utf-8') as currentFile:
        data = json.load(currentFile)
        for dict in data:
            bm25_answer_dict[dict["passage_idx"]] = dict["answer"]
```

```python
bm25_answer_dict
```

```python
correct = 0
for key, answer in bm25_answer_dict.items():
    true_answer = answer_dict.get(key)
    if answer == true_answer:
        correct += 1

print("The accuracy is: ", correct/len(bm25_answer_dict))
```
